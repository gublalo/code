**13. Todo parece indicar que las variables categ√≥ricas que quedan s√≠ son importantes para obtener la variable de salida. Convierte todas las variables categ√≥ricas que quedan a num√©ricas:**

**a) BusinessTravel**

**b) Department**

**c) EducationField**

**d) Gender**

**e) JobRole**

**f) MaritalStatus**

**g) Attrition**

```python
businessTavel_OH = pd.get_dummies(EmpleadosAttrition.BusinessTravel, prefix = 'BT')
EmpleadosAttrition.drop(['BusinessTravel'], axis = 1, inplace = True)
EmpleadosAttrition = pd.concat([EmpleadosAttrition, businessTavel_OH], axis = 1)

Department_OH = pd.get_dummies(EmpleadosAttrition.Department, prefix = 'D')
EmpleadosAttrition.drop(['Department'], axis = 1, inplace = True)
EmpleadosAttrition = pd.concat([EmpleadosAttrition, Department_OH], axis = 1)

EducationField_OH = pd.get_dummies(EmpleadosAttrition.EducationField, prefix = 'EF')
EmpleadosAttrition.drop(['EducationField'], axis = 1, inplace = True)
EmpleadosAttrition = pd.concat([EmpleadosAttrition, EducationField_OH], axis = 1)

Gender_OH = pd.get_dummies(EmpleadosAttrition.Gender, prefix = 'G')
Gender_OH.drop(['G_Female'], axis = 1, inplace = True)
EmpleadosAttrition.drop(['Gender'], axis = 1, inplace = True)
EmpleadosAttrition = pd.concat([EmpleadosAttrition, Gender_OH], axis = 1)

JobRole_OH = pd.get_dummies(EmpleadosAttrition.JobRole, prefix = 'JR')
EmpleadosAttrition.drop(['JobRole'], axis = 1, inplace = True)
EmpleadosAttrition = pd.concat([EmpleadosAttrition, JobRole_OH], axis = 1)

MaritalStatus_OH = pd.get_dummies(EmpleadosAttrition.MaritalStatus, prefix = 'MS')
EmpleadosAttrition.drop(['MaritalStatus'], axis = 1, inplace = True)
EmpleadosAttrition = pd.concat([EmpleadosAttrition, MaritalStatus_OH], axis = 1)

Attrition_OH = pd.get_dummies(EmpleadosAttrition.Attrition, prefix = 'A')
Attrition_OH.drop(['A_No'], axis = 1, inplace = True)
EmpleadosAttrition.drop(['Attrition'], axis = 1, inplace = True)
EmpleadosAttrition = pd.concat([EmpleadosAttrition, Attrition_OH], axis = 1)

EmpleadosAttrition.head()
```

**16. Crea una nueva variable llamada EmpleadosAttritionPCA formada por los componentes principales del frame EmpleadosAttritionFinal. Recuerda que el resultado del proceso PCA es un numpy array, por lo que, para hacer referencia a una columna, por ejemplo, la 0, puedes usar la instrucci√≥n EmpleadosAttritionPCA[:,0]).**

```python
pca = PCA()
EmpleadosAttritionPCA = pca.fit_transform(EmpleadosAttritionFinal)
EmpleadosAttritionPCA
```

**17. Agrega el m√≠nimo n√∫mero de Componentes Principales en columnas del frame EmpleadosAttritionPCA que logren explicar el 80% de la varianza, al frame EmpleadosAttritionFinal. Puedes usar la instrucci√≥n assign, columna por columna, llamando a cada unaC0, C1, etc., hasta las que vayas a agregar.**

```python
pca.explained_variance_ratio_ * 100
```

```python
np.sum(pca.explained_variance_ratio_[0] + pca.explained_variance_ratio_[1] + pca.explained_variance_ratio_[2]) * 100
```

```python
EmpleadosAttritionFinal = EmpleadosAttritionFinal.assign(C0 = EmpleadosAttritionPCA[:, 0])
EmpleadosAttritionFinal = EmpleadosAttritionFinal.assign(C1 = EmpleadosAttritionPCA[:, 1])
EmpleadosAttritionFinal = EmpleadosAttritionFinal.assign(C2 = EmpleadosAttritionPCA[:, 2])
EmpleadosAttritionFinal.head()
```
