```python
#a. Extraer la informaci√≥n del archivo.
import pandas as pd

countryVaccinations = pd.read_csv("/content/drive/MyDrive/ColabNotebooks/country_vaccinations.csv", parse_dates = ['date'])
countryVaccinations
```

```python
#b. Mostrar la estructura y tipos de datos de cada columna para identificar qu√© operaciones puedes realizar con cada una de ellas, 
#asegur√°ndote que las columnas con fechas sean del tipo datetime64. 
countryVaccinations.info()
```

```python
#c. Determinar la cantidad de vacunas aplicadas de cada compa√±√≠a (con base en c√≥mo lo reporta cada pa√≠s en la columna vaccines).
countryVaccinations['vaccines'].value_counts()
```

```python
#d. Obtener la cantidad de vacunas aplicadas en todo el mundo.
countryVaccinations['daily_vaccinations'].sum()
```

```python
#e. Calcular el promedio de vacunas aplicadas por pa√≠s.
countryVaccinations[['country', 'daily_vaccinations']].groupby('country').mean()
```

```python
#f. Determinar la cantidad de vacunas aplicadas el d√≠a 29/01/21 en todo el mundo.
countryVaccinationsR = countryVaccinations[countryVaccinations['date'] == pd.to_datetime('2021-01-29')]
countryVaccinationsR['daily_vaccinations'].sum()
```

```python
#g. Crear un dataframe nuevo denominado conDiferencias que contenga los datos originales y una columna derivada (diferencias) 
#con las diferencias de aplicaci√≥n entre las columnas daily_vaccionations y daily_vaccionations_raw.
conDiferencias = pd.DataFrame(countryVaccinations)
conDiferencias['diferencias'] = conDiferencias['daily_vaccinations'] - conDiferencias['daily_vaccinations_raw']
conDiferencias['diferencias']
```

```python
#h. Obtener el per√≠odo de tiempo entre el registro con fecha m√°s reciente y el registro con fecha m√°s antigua.
countryVaccinations['date'].max() - countryVaccinations['date'].min()
```

```python
#i. Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac) con la 
#cantidad de vacunas utilizadas cada d√≠a (usar la columna vaccines y separar por el car√°cter /).
conCantidad = pd.DataFrame(countryVaccinations)
conCantidad['canVac'] = conCantidad['vaccines'].str.split('/').str.len()
conCantidad['canVac']
```

```python
#j. Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020.
antes20 = countryVaccinations[countryVaccinations['date'] < pd.to_datetime('2020-12-20')]
antes20
```

```python
#k. Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer.
pfizer = countryVaccinations[countryVaccinations['vaccines'].str.contains('pfizer', case = False)]
pfizer
```

```python
#l. Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado 
#resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente.
with pd.ExcelWriter("/content/drive/MyDrive/ColabNotebooks/resultadosReto.xlsx") as pathFile:
  conDiferencias.to_excel(pathFile, sheet_name = "conDiferencias")
  conCantidad.to_excel(pathFile, sheet_name = "conCantidad")
  antes20.to_excel(pathFile, sheet_name = "antes20")
  pfizer.to_excel(pathFile, sheet_name = "pfizer")
```
