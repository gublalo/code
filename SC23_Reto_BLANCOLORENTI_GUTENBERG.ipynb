```python
#Crea DataFrames en pandas train.
import pandas as pd

train = pd.read_csv("/content/drive/MyDrive/ColabNotebooks/Train.csv")
train
```

```python
#Crea DataFrames en pandas train2.
import pandas as pd

train2 = pd.read_csv("/content/drive/MyDrive/ColabNotebooks/Train2.csv")
train2
```

```python
#Crea DataFrames en pandas test.
import pandas as pd

test = pd.read_csv("/content/drive/MyDrive/ColabNotebooks/Test.csv")
test
```

```python
#Crea DataFrames en pandas test2.
import pandas as pd

test2 = pd.read_csv("/content/drive/MyDrive/ColabNotebooks/Test2.csv")
test2
```

```python
#Conjunta de manera correcta la información, de tal forma que queden todos los registros juntos en un solo DataFrame 
#que contenga todas las columnas necesarias.
import pandas as pd

empleado = pd.concat([train, train2], ignore_index = True)
prueba = pd.concat([test, test2], ignore_index = True)
empleadoPrueba = pd.merge(empleado, prueba, on = "Employee_ID")
empleadoPrueba
```

```python
#Para todos los registros que no tienen un valor para un sistema operativo, ya sea de su computadora o de su dispositivo móvil, 
#asignar el valor "no registrado".
import pandas as pd

empleadoPrueba.loc[empleadoPrueba[' Computer_OS'].isnull(), ' Computer_OS'] = 'No Registrado'
empleadoPrueba.loc[empleadoPrueba[' Mobile_OS'].isnull(), ' Mobile_OS'] = 'No Registrado'
empleadoPrueba
```

```python
#El porcentaje de personas que usan Windows / Mac / Linux / no registrado como su sistema operativo en su computadora.
import numpy as np

total = 0

porcentajeCSO = empleadoPrueba.groupby(by=[" Computer_OS"]).agg({" Computer_OS":np.size})
total = int(porcentajeCSO.sum())
porcentajeCSO['Percent_OS'] = (porcentajeCSO[' Computer_OS'] / total) * 100
porcentajeCSO 
```

```python
#El porcentaje de personas que usan iOS / Android / No registrado como su sistema operativo en su dispositivo móvil.
import numpy as np

total = 0

porcentajeMSO = empleadoPrueba.groupby(by=[" Mobile_OS"]).agg({" Mobile_OS":np.size})
total = int(porcentajeMSO.sum())
porcentajeMSO['Percent_OS'] = (porcentajeMSO[' Mobile_OS'] / total) * 100
porcentajeMSO
```

```python
#La edad promedio de la gente que utiliza cada sistema operativo de escritorio.
import numpy as np

edadPromedio = empleadoPrueba.groupby(by=[" Computer_OS"]).agg({"Age":np.mean})
edadPromedio
```

```python
#El nivel educativo promedio de la gente que utiliza cada sistema operativo de escritorio.
import numpy as np

nivelEducativoPromedio = empleadoPrueba.groupby(by=[" Computer_OS"]).agg({"Education_Level":np.mean})
nivelEducativoPromedio
```

```python
#El número total de tickets generados por la gente con cada sistema operativo de escritorio.
import numpy as np

totalTickets = empleadoPrueba.groupby(by=[" Computer_OS"]).agg({" Computer_tickets":np.sum})
totalTickets
```

```python
#El resultado de cada solicitud de información a un archivo XLS[X], donde el resultado de cada consulta debe ir 
#en su propia hoja de trabajo. 
with pd.ExcelWriter("/content/drive/MyDrive/ColabNotebooks/SC23_Reto_BLANCOLORENTI_GUTENBERG.xlsx") as pathFile:
  empleadoPrueba.to_excel(pathFile, sheet_name = "personaTest")
  porcentajeCSO.to_excel(pathFile, sheet_name = "porcentajeCSO")
  porcentajeMSO.to_excel(pathFile, sheet_name = "porcentajeMSO")
  edadPromedio.to_excel(pathFile, sheet_name = "edadPromedio")
  nivelEducativoPromedio.to_excel(pathFile, sheet_name = "nivelEducativoPromedio")
  totalTickets.to_excel(pathFile, sheet_name = "totalTickets")
```
